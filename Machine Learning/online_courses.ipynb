```python
import pandas as pd
import numpy as np
```

```python
courses_df = pd.read_csv(filepath_or_buffer="Online_Courses.csv")
courses_df.head()
courses_df.columns
```

```python
data = {"Nombre":courses_df["Title"],"Categoria":courses_df["Category"],"URL":courses_df["URL"],"Descripcion":courses_df["Short Intro"],"Skills":courses_df["Skills"],
        "Instructores":courses_df["Instructors"],"Rating":courses_df["Rating"],"Viewers":courses_df["Number of viewers"]}
df = pd.DataFrame(data=data)
df[["Categoria","Descripcion","Skills","Instructores"]] = df[["Categoria","Descripcion","Skills","Instructores"]].fillna("Other")
df[["Rating","Viewers"]] = df[["Rating","Viewers"]].fillna("0")
df.head()
```

```python
df["Rating"] = df["Rating"].map(lambda x: float(x.replace("stars","")))
df
```

```python
import re
df["Viewers"] = df["Viewers"].map(lambda x: x.replace(",",""))
df["Viewers"] = df["Viewers"].map(lambda x: re.findall(pattern=r"\d+", string=x)[0])
df.head()
```

```python
df.to_csv(path_or_buf="courses_data.csv",sep=",",encoding='utf-8',index=False)
```

 ## Datos de Entrenamiento y Test


Tenemos que cambiar también como representamos las skills en el DataFrame.

```python
df["Skills"] = df["Skills"].map(lambda x: x.lower())
```

```python
df["Skills"] = df["Skills"].map(lambda x: x.split(","))
```

```python
skills = df["Skills"]

skills_copy = []
for i in range(len(skills)):
    lista_aux = []
    for elem in skills[i]:
        elem = re.sub('[^a-zA-Z]+', ' ',elem)
        elem = elem.lstrip()
        lista_aux.append(elem)
    skills_copy.append(lista_aux)
```

```python
df["Skills"] = skills_copy
df
```

```python
skills = df["Skills"]
```

```python
skills_unique = []
for lista_skills in skills:
    skills_unique += lista_skills
skills_unique = sorted(list(set(skills_unique)))
print(len(skills_unique))
number_of_skills = len(skills_unique)
skills_unique
```

## Creación Sintética de Datos

Tenemos que mezclar los datos de input y los datos de output para generar datos sintéticos.

```python
import random
```

```python
train_users = []
train_recomendations = []

output_df = pd.DataFrame(output)

for i in range(25000):
    # Creamos un usuarios y una recomendación vacía:
    user = np.zeros(shape=(number_of_skills))
    recomendation = np.zeros(shape=(8092))
    # Generamos un conjunto de skills aleatorias para el usuario
    random_list = random.sample(range(1, number_of_skills), np.random.randint(low=1,high=100))
    user[random_list] = 1
    train_users.append(user)
    # Creamos los cursos para ese usurio:
    courses_for_skills = []
    for random_skill in random_list:
        
        rows = output_df.index[output_df[random_skill] == 1].tolist()
        courses_for_skills += rows
        
    recomendation[courses_for_skills] = 1
    train_recomendations.append(recomendation)
```

```python
np.save("users.npy",np.array(train_users))
np.save("recomendations.npy",np.array(train_recomendations))
```
